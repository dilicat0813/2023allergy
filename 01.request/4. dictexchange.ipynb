```python
import pandas as pd
from konlpy.tag import Okt

datas = df['allergy2']

# allergy_dict 미리 정의
allergy_dict = {1: ['알류', '계란', '달걀', '난류','메추리알','메추리','추리알','난각','날류','난백','난황','난각','게란'], 
                2: ['우유','유우', '탈지분유','유유','분유','유청','카제인','우류'], 
                3: '메밀', 
                4: ['땅콩', '탕콩'],
                5: ['대두','두부','대듀','콩','적두','유부'], 
                6: ['밀','밀가루'], 
                7: '고등어', 
                8: '게', 
                9: ['새우'], 
                10: ['돼지고기','괘지','괘지고기','돈','돼지','돼고기','돈창', '돈혈', '돈지방','괘'],
                11: ['복숭아'], 
                12: ['토마토', '토마투','토마도','토마','토미토'],
                13: ['아황산', '아황산나트륨', '이황산', '이산화', '아산화', '이산화황','아산화황','이산화항','이황산류'],
                14: ['호두','피칸','호도'], 
                15: ['닭고기', '계육','닭','닭괴'], 
                16: ['쇠고기', '우육', '소고기', '소뼈','쇠구기'], 
                17: ['오징어', '오지엉','오지','지엉'],
                18: ['조개류','굴','홍합','조개','백합','바지락'], 
                19: '잣',
                20: ['아몬드','페닐알라닌','헤이즐넛','참깨']}

allergy_nums = []
for data in datas:
    data = str(data)
    okt = Okt()
    words = okt.pos(data)
    allergy_columns = []

    for word, tag in words:
        if tag == 'Noun':
            for allergy_key, allergy_value in allergy_dict.items():
                if isinstance(allergy_value, list):
                    if word in allergy_value:
                        allergy_columns.append(allergy_key)
                        break
                else:
                    if word == allergy_value:
                        allergy_columns.append(allergy_key)
                        break
        
    allergy_num = ','.join(str(x) for x in allergy_columns) # 문자열로 변환
    allergy_nums.append(allergy_num)

df['allergy_num'] = allergy_nums # DataFrame에 열 추가
print(df)
```
